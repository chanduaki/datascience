```python
import matplotlib.pyplot as plt
```

## Line Plot

```python
 x = [1,2,3,4,5]
 y = [2,4,6,8,10]
```

```python
plt.plot(x,y)
```

```python
# Lets add some beautification to it 
plt.plot(x,y)
plt.xlabel("X Data")
plt.ylabel("Y Data")
plt.title("Basic Line Plot")
plt.show()
```

```python
# Lets create some customised line plot 
plt.plot(x,y,color="red", linestyle="--", marker='o', linewidth=3,markersize=9)
plt.grid(True)
```

## Bar Plot

```python
categories=['A','B','C','D','E']
values=[5,7,3,8,6]
```

```python
plt.bar(categories,values,color="blue")
```

## Pie Chart

```python
labels=['A','B','C','D']
sizes=[30,20,40,10]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0) ##move out the 1st slice
```

```python
plt.pie(sizes,labels=labels,colors=colors)
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%")
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%", explode=explode)
```

```python
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%", explode=explode, shadow=True)
```
